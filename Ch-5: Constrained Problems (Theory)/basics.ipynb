# Ch-5 Constrained Problems - Theory

This chapter is all about finding a min of objective function, subject to certain restrictions on the variables ***Constraints***. The solution is restricted to feasible sets, defined by the constraints. These constraints can be equations or inequalities, which define regions, under which the solution must lie.

#### Analysis b/w Unconstrained and Constrained optimization

In unconstrained, we are free to search the entire domain of objective function, however, in constrained optimization, the solution is restricted to feasible sets (defined by the additional constraints equation/ inequalities). This makes the process more complex, and involves additional mathematical tools, such as ***Langrange Multipliers***, to handle constraints effectively. <br>

Similar to unconstraint optimization, the constraints optimization also has:
1. **Necessary Conditions**: 1st order conditions are explained by KKT conditions, often involves langrange multipliers. 
2. **Sufficient Conditions**: 2nd order conditions, involve analysis of curvature around the optimal value, which can be obtained via Hessian of langrangian. If a point satisfies KKT conditions, and Hessian of langrangian is positive def **on the feasible direction space**, then the point is local min. (2nd order sufficient condition)
